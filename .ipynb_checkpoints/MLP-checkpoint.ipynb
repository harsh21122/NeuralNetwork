```python
import random
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore") 
```

```python
from NeuralNetwork import MLP
def accuracy(y_true, y_predicted):
    count = 0
    for i in range(len(y_true)):
        if(y_true[i] == y_predicted[i]):
            count = count + 1
    return (count/len(y_true))

def classwise_accuracy(Y_true, Y_pred):
    count = np.zeros((10,10))
    for itr in range(len(Y_true)):
        true= Y_true[itr][0]
        pred = Y_pred[itr]
        count[true][pred] = count[true][pred] + 1
    count = count.astype(int)
    return count.diagonal()/count.sum(axis=1)
```

```python
# !pip install idx2numpy
# Downloaded MNIST dataset from http://yann.lecun.com/exdb/mnist/

import idx2numpy
y_train = idx2numpy.convert_from_file('./dataset/train-labels-idx1-ubyte')
x_train = idx2numpy.convert_from_file('./dataset/train-images-idx3-ubyte')
x_test = idx2numpy.convert_from_file('./dataset/t10k-images-idx3-ubyte')
y_test = idx2numpy.convert_from_file('./dataset/t10k-labels-idx1-ubyte')
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1]*x_train.shape[2])
y_train = y_train.reshape(-1, 1)
print(x_train.shape, y_train.shape)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2])
y_test = y_test.reshape(-1, 1)
print(x_test.shape, y_test.shape)
```

```python
#Normalize


x_train = (x_train - np.min(x_train)) / (np.max(x_train) - np.min(x_train))
x_test = (x_test - np.min(x_test)) / (np.max(x_test) - np.min(x_test))

print(x_train.shape, x_test.shape)
```

```python
np.max(x_train)
```

```python
np.unique(y_train, return_counts = True)
```

<h3>MLP training with Sigmoid

```python
nn = MLP(learning_rate = 0.5, activation_function = ['sigmoid'],  
                 hidden_layers = [128], optimizer = 'gradient_descent', batch_size = 60000, epochs = 500, seed = 1)

nn.fit(x_train, y_train)
nn.empty_variables()

#Save
filename = './model/nn_sigmoid_config_1.sav'
joblib.dump(nn, filename)
```

```python
#Load
filename = './model/nn_sigmoid_config_1.sav'
nn = joblib.load(filename)

plt.plot(np.arange(1, nn.epochs + 1, 1), nn.loss_per_epoch)
plt.xlabel("Number of epochs")
plt.ylabel("Loss")
y_pred = nn.predict(x_train)
print("Train Accuracy: ", accuracy(y_train, y_pred))
y_pred = nn.predict(x_test)
print("Test Accuracy: ", accuracy(y_test, y_pred))
print("Test Classwise Accuracy: ", classwise_accuracy(y_test, y_pred))
plt.show()
```
